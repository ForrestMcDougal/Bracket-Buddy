```python
import numpy as np
import sklearn
import pandas as pd
import tensorflow as tf
from keras.models import Sequential
from keras.utils import to_categorical
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv('./data/ml_data/model_data.csv', index_col=False)
# This is the cleaned data merged from SQL
```

```python
df = df[['home_points', 'away_points', 'a_AdjTempo', 'a_AdjOE', 'a_eFG_Pct_O', 'a_eFG_Pct_D', 'a_OFF_3PT', 'a_FG2Pct', 'a_FG3Pct', 'a_FTPct', 'a_BlockPct', 'a_OppFG2Pct', 'a_OppFG3Pct', 'a_OppFTPct', 'a_F3GRate', 'a_StlRate', 'a_OppStlRate', 'a_TO_Pct_O', 'a_TO_Pct_D', 'a_OFF_2PT', 'h_AdjTempo', 'h_AdjOE', 'h_eFG_Pct_O', 'h_eFG_Pct_D', 'h_OFF_3PT', 'h_FG2Pct', 'h_FG3Pct', 'h_FTPct', 'h_BlockPct', 'h_OppFG2Pct', 'h_OppFG3Pct', 'h_OppFTPct', 'h_F3GRate', 'h_StlRate', 'h_OppStlRate', 'h_TO_Pct_O', 'h_TO_Pct_D', 'h_OFF_2PT', 'hca']]
```

```python
X1 = df.drop(['home_points','away_points'], axis=1).values
y1 = df[['home_points','away_points']].values
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, random_state=42)

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X1_train)
# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X1_train)
X_test_scaled = X_scaler.transform(X1_test)
```

```python
print(X_train_scaled.shape, X_test_scaled.shape, y1_train.shape, y1_test.shape)
```

```python
# Pick a random value for units and activation whatever!
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=37))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=2))
```

```python
# YMMV
model.compile(optimizer='adam',
              loss='mse',
              metrics=['accuracy'])
```

```python
model.fit(
    X1_train,
    y1_train,
    epochs=200,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X1_test, y1_test, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model.save("lw_model_trained.h5")
```

```python
y_pred = model.predict(X1_test)
# Predicted Y values using the model

y_val_df = pd.DataFrame(y_pred)
# Put them into a dataframe for easier munging
y_val_df['home_win'] = y_val_df[0] > y_val_df[1]
y_val_df['overunder'] = y_val_df[0] + y_val_df[1]
y_val_df['spread'] = y_val_df[0] - y_val_df[1]
# Add a column for if the home team won in the prediction
y_val_df.head(10)
```

```python
y_test_df = pd.DataFrame(y1_test)
y_test_df['home_win'] = y_test_df[0] > y_test_df[1]
# Add a column for home win to the test data
y_test_df['overunder'] = y_test_df[0] + y_test_df[1]
y_test_df['spread'] = y_test_df[0] - y_test_df[1]
y_test_df.head(10)
# Actual game scores
```
