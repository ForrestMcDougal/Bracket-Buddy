# Scraping KenPom HCA page

```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
import os
import glob
from config import user, pwd
import numpy as np
```

```python
def init_browser():
    # Mac-specific browser init
    executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
    browser = Browser('chrome', **executable_path, headless=False)
    return browser
```

```python
url = 'https://kenpom.com'
password = pwd
email = user
browser = init_browser()
browser.visit(url)
```

```python
browser.find_by_name('email').fill(email)
browser.find_by_name('password').fill(password)
browser.find_by_name('submit').first.click()
```

```python
url2 = 'https://kenpom.com/hca.php'
browser.visit(url2)
```

```python
# Scrape the page
soup = BeautifulSoup(browser.html, 'lxml')
browser.quit()
headline_list = [] # Store the headlines

# Find the table
table = soup.find_all('table')[0] 
```

```python
df = pd.read_html(str(table))
```

# Getting unique team names from all the Kenpom data

```python
df1 = pd.DataFrame()
# Change directory
data_loc = "./data/cleaned"#input("Enter the path for your download directory: ")
try:
    os.chdir(data_loc)
except FileNotFoundError:
    print("Please change the directory to the location of your results files")

# List all files and directories in current directory
files = glob.glob((data_loc + "/*.csv").split("/")[-1])
print(files)
```

```python
file_count = 0
for item in files:

    xl = pd.read_csv(item)
    df1 = df1.append(xl, ignore_index=True, sort=False)
    # Get the portion of the filename
    file_count += 1
    # Load a sheet into a DataFrame by name: df1
```
